### Постановка задачи <a class="anchor"  id="chapter1.1"></a>

Предстоит работать с псевдо-реальными (реальные данные, в которые добавлена синтетическая составляющая) данными о стартапах, функционировавших в период с 1970 по 2018 годы, и предсказать, какие из них закроются.

__Оценка качества предсказаний__

Метрикой качества является `F1-Score`. `F1-мера` вычисляется на основе точности `p` и полноты `r`. Точность - это отношение `true positives (tp)` ко всем `predicted positives (tp + fp)`. Полнота это отношение `true positives` ко всем `actual positives (tp + fn)`.

Метрика `F1` является гармоническим средним между `Recall` и `Precision`.

Хорошая модель при обучении будет максимизировать `Recall` и `Precision` одновременно. Таким образом, умеренно хорошие показатели по обоим параметрам будут предпочтительнее, чем чрезвычайно хорошие показатели по одному и плохие по другому. Функция для вычисления метрики - https://scikit-learn.org/stable/modules/generated/sklearn.metrics.f1_score.html.

Минимальное значение метрики `0.71186`.

Значение метрики на случайной модели `0.15270`.

### Вспомогательные функции <a class="anchor"  id="chapter2.3"></a>

```python
def show_duplicates(df: pd.DataFrame, column = ''):
    """
    Выводит сведения о количестве явных дубликатов в столбце `column` для данных `df`
    Если столбце `column` не задан то выводятся явные дубликаты по совокупности всех столбцов

    Параметры
    =========
        `df` : DataFrame
        `column` : string, не обязательный
                   Название столбца
    """
    tmp_df = df[column] if column else df
    info = f"столбце `{column}`"  if column else 'данных'

    total = tmp_df.shape[0]

    print(f"Количество явных дубликатов в {info}: {tmp_df.duplicated().sum()} ({tmp_df.duplicated().sum() / total:.2%})")
```

```python
def show_na(df: pd.DataFrame, column: str = None, count: int=2):
    '''
    Выводит количество пустых данных во фрейме или для конкретного параметра
    Если количество больше нуля, то выводятся `count` строк с нулевыми  значениями

    Параметры
    =========

    `df`: дата фрейм
    `column`: название параметра (необязательно)
    `count`: количество первых пустых строк (необязательно)
    '''
    tmp_df = df if column == None else df[column]
    column_info = '' if column == None else f' ({column})'

    count_na = tmp_df.isna().sum()

    total = tmp_df.shape[0]

    print(f'Количество пустых данных{column_info}: {count_na} ({count_na/total: .2%})')

    if count_na > 0:
        display(df[tmp_df.isna()].head(count))
```

```python
def show_corr(corr: pd.DataFrame, min_corr_value=0.6, figsize=(12, 10), title_corr=''):
    """
    Визуализирует коэффициенты корреляции для данных `df` и минимального значения `min_corr_value`

    Параметры
    ---------

    corr : DataFrame
         матрица корреляции
    min_corr_value : float, значение по умолчанию 0.6
                     минимальное значение для отсеивания значений корреляции
    figsize : (width, height), по умолчанию (12, 10)
              размер фигуры
    """
    mask = np.triu(np.ones_like(corr))

    plt.figure(figsize=figsize)

    sns.heatmap(
        corr[corr.abs() >= min_corr_value],
        annot=True,
        mask=mask,
        linewidth=.5,
        fmt=".2f",
        cmap='binary')

    plt.title(f"Коэффициенты коррелцяии: {title_corr}")
```

```python
def show_numeric_distribution(df, column, bins=10):
    """
    Выводит график распределения данных для df[column], ящик с усами и статистику данных

    Параметры
    =========
    df - DataFrame
    column - названия столбца в df
    bins - по умолчанию 10, количество разбиений для распределения данных
    """
    _, axs = plt.subplots(2, 1, sharex=True, gridspec_kw={'height_ratios': [3, 1]})

    # статистика данных
    tmp_df = df[column].describe().apply(round, ndigits=2)

    table = axs[0].table(cellText=tmp_df.to_numpy().reshape((-1,1)), rowLabels=tmp_df.index, loc='right', bbox=[1.1,0,.2,1,])

    # гистограмма
    sns.histplot(data=df, x=column, ax=axs[0], bins=bins, kde=True)
    axs[0].set_ylabel('Количество записей, шт.')
    axs[0].set_title(f"Распределение данных параметра: {column}")

    # ящик с усами
    sns.boxplot(data=df, x=column, ax=axs[1])

    tmp_df = df[column].describe()
    irq = tmp_df['75%'] - tmp_df['25%']
    irq_min = max(tmp_df['25%'] - 1.5 * irq, tmp_df['min'])
    irq_max = min(tmp_df['75%'] + 1.5 * irq, tmp_df['max'])
    tmp_df = df.query(f"{column} < @irq_min or {column} > @irq_max")
    if tmp_df.shape[0] > 0:
        print(f"Доля выбросов: {tmp_df.shape[0] / df.shape[0]:.4%}")


def to_pie_value(pct, total):
    '''
    Функция используется для отображения абсолютных значений и процентов в диаграмме типа круг (pie)

    Параметры:
    ==========

    pct - значение в процентах
    total - общие количество
    '''
    absolute = int(np.round(pct/100.* total))
    return f"{pct:.1f}%\n({absolute:_d})".replace('_', ' ')

def show_categorical_distribution(df, column):
    """
    Выводит распределение данных для категориальных данных

    Параметры:
    =========

    df - DataFrame
    column - название столбца в df
    """
    tmp_df = df[column].astype('string').fillna('- пустое значение -').value_counts()
    total = tmp_df.sum()

    explode = [0.02] * tmp_df.shape[0]

    plt.figure(figsize=(12,8))
    wedges, texts, autotexts = plt.pie(tmp_df, autopct=lambda v: to_pie_value(v, total), explode=explode)

    plt.legend(wedges, tmp_df.index,
          title=column,
          loc="center left",
          fontsize=14,
          bbox_to_anchor=(1, 0, 0.5, 1))

    plt.setp(autotexts, size=12, weight="bold")
    plt.title(f"Распределение данных: {column}")



def show_distribution(df, column, bins=10):
    """
    В зависимости от типа данных df[column]

    Если числовой, то выводит график распределения данных для df[column], ящик с усами и статистику данных

    Если категориальный, то выводит распределение данных для категориальных данных

    Параметры
    =========
    df - DataFrame
    column - названия столбца в df
    bins - по умолчанию 10, количество разбиений для распределения данных
    """

    if pd.api.types.is_numeric_dtype(df[column].dtype):
        show_numeric_distribution(df, column, bins=bins)
    else:
        show_categorical_distribution(df, column)
```

```python
def drop_column_duplicates(df, column):
    """
    Возвращает данные с удаленными строками с явными дубликатами в столбце `column`

    Параметры
    =========
        df : DataFrame
             данные
        column : string
                 столбце в котором нужно искать явные дубликаты
    """
    # индексы с дублями
    drop_index = df[df[column].duplicated()].index
    # удаляем дубликаты
    return df.drop(index=drop_index)

def drop_irq_value(df, column):
    """
    Возвращает данные с удаленными выбросами по +/-1.5 * irq выбросами в столбце `column`

    Параметры
    =========
        df : DataFrame
             Данные
        column : string
                 название столбца
    """

    tmp_df = df[column].describe()
    irq = tmp_df['75%'] - tmp_df['25%']
    irq_min = max(tmp_df['25%'] - 1.5 * irq, tmp_df['min'])
    irq_max = min(tmp_df['75%'] + 1.5 * irq, tmp_df['max'])

    # индексы с выбросами
    drop_index = df.query(f"{column} < @irq_min or {column} > @irq_max").index
    return df.drop(index=drop_index)
```

Стоит отметить, что в данных имеется зависимость между данными: например, все категориальные данные связанные со странами, регионами, штатами и городами имеют весьма существенную корреляцию данных.

Между датой основания и длительностью стартапа имеется линейная зависимость. Причем точки, которые выходят за линию являются закрытыми стартапами. Данное свойство использовалось для аналитического решения.

В связи с этим в качестве основных параметров для моделей предсказания `status` использовать:
- `country_code`;
- `funding_total_usd`;
- `funding_rounds`;
- `founded_at`;
- `founded_at_year`;
- `first_funding_at`;
- `last_funding_at`;
- `funding_at_dt`;
- `lifetime`.

В ходе анализа лучше модели выявлены следующие важные признаки влияющие на предсказание статуса стартапа:
- длительность функционирования стартапа;
- дата основания стартапа;
- код страны.

В качестве дальнейших действий по улучшению моделей рекомендуется:
- исключить зависимость в данных между датой основания и функционирования стартапа, например, за счет добавления дополнительных данных. Например, для функционирующего стартапа можно указать несколько строк с разной длительностью функционирования, т.к. стартап функционировал 3 года, то оже функционировал и 2 года и меньше. Добавление таких данных может помочь разрушить зависимость между данными;
- убрать не явные дублирования в данных категорий;
- использовать методы борьбы с дисбалансом в данных путем генерирования дополнительных данных и т.п.;
- для лучшей модели можно осуществить оптимизацию гиперпараметров.
- исключить из модели параметр `lifetime` и построить модель на собственных характеристиках стратапа.
