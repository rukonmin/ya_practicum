# Проверка гипотез данных сервиса аренды самокатов

## Общие сведения

### Цель исследования (постановка задачи)

Вы аналитик популярного сервиса аренды самокатов GoFast. Вам передали данные о некоторых пользователях из нескольких городов, а также об их поездках. Проанализируйте данные и проверьте некоторые гипотезы, которые могут помочь бизнесу вырасти.

Чтобы совершать поездки по городу, пользователи сервиса GoFast пользуются мобильным приложением. 

Сервисом можно пользоваться:

- без подписки `free`:
   - абонентская плата отсутствует;
   - стоимость одной минуты поездки — 8 рублей;
   - стоимость старта (начала поездки) — 50 рублей;
- с подпиской `ultra`:
   - абонентская плата — 199 рублей в месяц;
   - стоимость одной минуты поездки — 6 рублей;
   - стоимость старта — бесплатно.
   
Необходимо провести исследовательский анализ данных:

- Частота встречаемости городов
- Соотношение пользователей с подпиской и без подписки
- Возраст пользователей
- Расстояние, которое пользователь преодолел за одну поездку
- Продолжительность поездок

Рассчитать значения выручки:

- найдите суммарное расстояние, количество поездок и суммарное время для каждого пользователя за каждый месяц.
- помесячная выручка, которую принёс каждый пользователь. Продолжительность каждой поездки в каждой строке исходного датафрейма для подсчёта стоимости округляется до следующего целого числа: например, значения 25.3, 25.5 и 26.0 должны быть преобразованы к 26.

Проверить гипотезы:

- Важно понять, тратят ли пользователи с подпиской больше времени на поездки? Если да, то пользователи с подпиской могут быть «выгоднее» для компании. Проверьте гипотезу. Используйте исходные данные о продолжительности каждой сессии — отдельно для подписчиков и тех, у кого нет подписки.
- Расстояние одной поездки в 3130 метров — оптимальное с точки зрения износа самоката. Можно ли сказать, что среднее расстояние, которое проезжают пользователи с подпиской за одну поездку, не превышает 3130 метров? Проверьте гипотезу и сделайте выводы.
- Проверьте гипотезу о том, будет ли помесячная выручка от пользователей с подпиской по месяцам выше, чем выручка от пользователей без подписки. Сделайте вывод.
- Представьте такую ситуацию: техническая команда сервиса обновила сервера, с которыми взаимодействует мобильное приложение. Она надеется, что из-за этого количество обращений в техподдержку значимо снизилось. Некоторый файл содержит для каждого пользователя данные о количестве обращений до обновления и после него. Какой тест вам понадобился бы для проверки этой гипотезы?

Осуществить предсказания:

- Отделу маркетинга GoFast поставили задачу: нужно провести акцию с раздачей промокодов на один бесплатный месяц подписки, в рамках которой как минимум 100 существующих клиентов должны продлить эту подписку. То есть по завершении периода действия подписки пользователь может либо отказаться от неё, либо продлить, совершив соответствующий платёж. ### Вспомогательные методы

````python
def boxplot_with_hist(df, column, figsize=(10, 5), title='', column_title='', bins=10):
    describe = df[column].describe()
    
    irq = describe['75%'] - describe['25%']
    irq_min = max(describe['min'], describe['25%'] - 1.5 * irq)
    irq_max = min(describe['max'], describe['75%'] + 1.5 * irq)
    
    _, axs = plt.subplots(1, 2, figsize=figsize)
    axs[0].sharey(axs[1])
    
    if len(column_title) == 0:
        column_title = column
        
    axs[0].set_ylabel(column_title)
    axs[1].set_xlabel('Количество данных, шт.')

    df.boxplot(column, ax=axs[0])
    df[column].hist(edgecolor='b', ax=axs[1], orientation='horizontal', bins=bins)
    
    axs[1].axhline(irq_min, color='r', linestyle=':')
    _, x = axs[1].get_xlim()
    axs[1].text(x + 10, irq_min, f"{irq_min:_.2f}", color='r', fontweight='bold', fontsize='12')
    
    axs[1].axhline(irq_max, color='r', linestyle=':')
    axs[1].text(x + 10, irq_max, f"{irq_max:_.2f}", color='r', fontweight='bold', fontsize='12')
    
    title = title if len(title) else f"Распределение данных {column}"
    plt.suptitle(title)

def show_barh_values(ax):
    for p in ax.patches:
        _, x = ax.get_xlim()
        ax.text(x * 1.005, p.get_y() + p.get_height() / 2, f"{p.get_width():_d}")
    return ax
        
def corr_to_text(value):
    value = abs(value)
    if 0 <= value < 0.3:
        return 'очень слабая'
    elif 0.3 <= value < 0.5:
        return 'слабая'
    elif 0.5 <= value < 0.7:
        return 'средняя'
    elif 0.7 <= value < 0.9:
        return 'высокая'
    elif 0.9 <= value <= 1:
        return 'очень высокая'
````

## Исследовательский анализ данных

### Вспомогательные методы

````python
def plot_bar_per_user(df, column, user_count=3, seed=1231, title=''):
    """
    Визуализация числовых данных для заданного количества пользователей
    
    Параметры:
        `df` - данные, 
        `column` - анализируемый параметр
        `user_count=3` - количество различных пользователей
        `seed=1231` - фиксируемая случайность
        `title=''` - текст заголовка
    """

    # фиксируем случайность
    rng = np.random.default_rng(seed)

    # для каждого пользователя рисуем его расстояния за одну поездку
    for user_id in rng.choice(df.user_id.unique(), user_count):
        user_data = df.query('user_id == @user_id')

        ax = user_data.plot.bar(
            y=column, 
            figsize=(10, 5),
            title=f"Пользователь ({user_id=})\n{title}")

        ax.set_xlabel('Порядковый номер поездки, шт.')
        ax.set_ylabel(title)

        # среднее значение
        mean_data = user_data[column].mean()
        ax.axhline(mean_data, color='red', linestyle='--')

        _, x = ax.get_xlim()
        ax.text(x * 1.01, mean_data, f"среднее = {mean_data:_.2f}", color='r', fontweight='bold', fontsize='12')
````

## Проверка гипотез

Шаг 6. Проверка гипотез

Продакт-менеджеры сервиса хотят увеличить количество пользователей с подпиской. Для этого они будут проводить различные акции, но сначала нужно выяснить несколько важных моментов.

### Гипотеза о трате больше времени на поездки пользователей с подпиской

_Описание задачи_

Важно понять, тратят ли пользователи с подпиской больше времени на поездки? 

Если да, то пользователи с подпиской могут быть «выгоднее» для компании. 

Проверьте гипотезу. 

Используйте исходные данные о продолжительности каждой сессии — отдельно для подписчиков и тех, у кого нет подписки.

- Гипотеза $H_0$ - пользователи с подпиской и без тратят в среднем одинаковое время на поездки.
- Гипотеза $H_1$ - пользователи с подпиской тратят в среднем больше времени на поездки чем без подписки.

1. Есть две независимые генеральные выборки:
   - пользователи с подпиской (`ultra_all_data`);
   - пользователи без подписки (`free_all_data`).
2. Выборочное среднее нормально распределено в соответствии с ЦПТ и тем фактом, что у нас достаточно большое количество значений.
3. предполагаем, что дисперсия разная, так как количество значений не равно.

````python
ultra_all_data.duration.hist(alpha=0.7)
ax = free_all_data.duration.hist(alpha=0.7)

ax.set_ylabel('Количество данных, шт.')
ax.set_xlabel('Длительность, минут')

ax.legend(['ultra', 'free'])

plt.suptitle('Распределение данных по длительности\nв зависимости от типа подписки')
````

Судя по графикам распределения, можно предположить, что дисперсия у них одинаковая.

Форма распределения схоже с нормальным распределением.

Случчай с не равной дисперсией.

````python
alpha = 0.01

# Примечание. При duration < 1 имеются аномальные данные
results = st.ttest_ind(
    ultra_all_data.duration, 
    free_all_data.duration, 
    equal_var=False, 
    alternative='greater')

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
````

Случай с равной дисперсией

````python
alpha = 0.01

# Примечание. При duration < 1 имеются аномальные данные
results = st.ttest_ind(
    ultra_all_data.duration, 
    free_all_data.duration, 
    equal_var=True, 
    alternative='greater')

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
````

_Вывод_

Есть основания полагать, что пользователи с подпиской тратят больше времени чем без.

### Гипотеза о среднем расстоянии за одну поезду для пользователя с подпиской

_Описание задачи_

Расстояние одной поездки в 3130 метров — оптимальное с точки зрения износа самоката. 

Можно ли сказать, что среднее расстояние, которое проезжают пользователи с подпиской за одну поездку, не превышает 3130 метров? 

Проверьте гипотезу и сделайте выводы.

- Гипотеза $H_0$ - среднее расстояние, которое проезжает пользователь с подпиской, равно 3130 метров.
- Гипотеза $H_1$ - среднее рассмтояние, которое проезжает пользователь с подпиской, больше 3130 метров..

1. Есть генеральная выборка: пользователи с подпиской (`ultra_all_data`).
2. Выборочное среднее нормально распределено в соответствии с ЦПТ и тем фактом, что у нас достаточно большое количество значений.

````python
alpha = 0.01

m = 3_130

results = st.ttest_1samp(
    ultra_all_data.distance, 
    m, 
    alternative='greater')

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
````

### Минимальное количество промокодов для рассылки с вероятностью не выполнить план ~5%

_Описание задачи_

Отделу маркетинга `GoFast` поставили задачу: нужно провести акцию с раздачей промокодов на один бесплатный месяц подписки, в рамках которой как минимум 100 существующих клиентов должны продлить эту подписку. То есть по завершении периода действия подписки пользователь может либо отказаться от неё, либо продлить, совершив соответствующий платёж.

Эта акция уже проводилась ранее и по итогу выяснилось, что после бесплатного пробного периода подписку продлевают `10%` пользователей. 

Выясните, какое минимальное количество промокодов нужно разослать, чтобы вероятность не выполнить план была примерно `5%`.

Подберите параметры распределения, описывающего эту ситуацию, постройте график распределения и сформулируйте ответ на вопрос о количестве промокодов.

````python
# минимальное число подписчиков продлившие бесплатную подписку
min_user_counts = 100

# имеется схема Бернулли с вероятностью успеха 0.1
# схема Бернулли повторяется n раз
# таким образом имеем биноминальное распределение
# математическое ожидание - n * p
# дисперсия - n * p * (1−p)
# P(k успехов из n попыток)=C_n^k * ⋅p^k * (1−p)^{n−k}
# F(x < 100) = 0.05

# для решения задачи можно воспользоваться аппроксимацией биноминального 
# 1. распределения распределением Пуассона P(n*p)
# 2. нормальным распределением - N(n * p, n * p * (1−p))

# Так как по условии задачи надо оценить количество не меньше 100, 
# то лучше всего подойдет аппроксимация нормальным распределением

# N(n * p, n * p * (1−p))

n = min_user_counts
p = 0.1

flag = True
while flag:
    m = n * p
    sigma = (n * p * (1 - p)) ** 0.5   
    
    distr = st.norm(m, sigma)
    x = distr.ppf(0.05)
    if x >= min_user_counts:
        flag = False
    else:
        n = n + 1

print(f"Необходимо разослать минимальное количество промокодов: {n}")

# рассчитаем параметры аппроксимации
m = n * p
sigma = (n * p * (1 - p)) ** 0.5
min_value = m - 3 * sigma
max_value = m + 3 * sigma

print(f"{0} <= {min_value=} ... {max_value=} <= {n=}")

# найдём суммарную ошибку аппроксимации:
distr = st.norm(m, sigma)
error_values = []
for i in range(n + 1):
    error = abs(st.binom.pmf(i, n, p) - (distr.cdf(i + 0.5) - distr.cdf(i - 0.5)))
    error_values.append(error)
 
print('Средняя разница вероятностей на значение:', sum(error_values) / len(error_values))
print('Максимальная разница вероятностей:', max(error_values))

distr = []
for k in range(0, n + 1):
    distr.append(st.binom.pmf(k, n, p))

plt.figure(figsize=(15, 6))
plt.bar(range(0, n + 1), distr)

plt.ylabel('Вероятность')
plt.xlabel(f"Принимаемые целые значения от 0 до {n}")

plt.title(f"Биноминальное распределение с {n=} и {p=}")
````

Рассмотрим как поведет себя апроксимация через распределение Пуасона

````python
n = min_user_counts
p = 0.1

flag = True
while flag:
    m = n * p
    
    x = st.poisson.ppf(0.05, m)
    if x >= min_user_counts:
        flag = False
    else:
        n = n + 1

print(f"Необходимо разослать минимальное количество промокодов: {n}")

# рассчитаем параметры аппроксимации
binom_probs = []
poisson_probs = []
for k in range(n + 1):
    # рассчитываем значения вероятности для биномиального распределения
    # и распределения Пуассона
    binom_prob = st.binom.pmf(k, n, p)
    poisson_prob = st.poisson.pmf(k, n*p)
    # добавляем в соответствующие списки
    binom_probs.append(binom_prob)
    poisson_probs.append(poisson_prob)
 
# рисуем графики
plt.figure(figsize=(15, 5))
plt.plot(range(n + 1), binom_probs, 'bo',
         label=f'Биномиальное распределение\nс n={n} и p={p}')
plt.plot(range(n + 1), poisson_probs, 'rd',
         label='Аппроксимация\nраспределением Пуассона')
plt.xlabel(f'k: количество успехов\n для B{n, p} / событий для Pois({n*p})')
plt.ylabel('Вероятность k')
plt.legend()
plt.xticks(range(n + 1))
plt.title('Сравнение распределений')
plt.show()

# найдем суммарную ошибку аппроксимации для соответствующих значений:
error_values = []
for i in range(n + 1):
    error = abs(binom_probs[i] - poisson_probs[i])
    error_values.append(error)
 
print('Средняя разница вероятностей на значение:', sum(error_values) / len(error_values))
print('Максимальная разница вероятностей:', max(error_values))


# Построим распределение
distr = []
for k in range(0, n + 1):
    distr.append(st.binom.pmf(k, n, p))

plt.figure(figsize=(15, 6))
plt.bar(range(0, n + 1), distr)
plt.xlabel(f"Принимаемые значения от 0 до {n}")
plt.ylabel('Вероятность k')
plt.title(f"Биноминально распределение с {n=} и {p=}")
plt.show()
````
